```python
import pandas as pd
import os
from dotenv import load_dotenv
import sqlalchemy
from sqlalchemy import create_engine
import time
import numpy as np

load_dotenv()
```

```python
engine = create_engine(os.getenv('DATABASE_URL1'))
```

```python
start_time = time.time()

query = '''select ch.id as cohort_id, ch.scientific_name, ch.local_name, trm.latitude, trm.longitude, trm.altitude from respi_cohort ch join respi_tree_measurement trm on trm.cohort_id=ch.id;'''
query1 = '''select sp.id as species_id, sp.scientific_name, sp.local_name, trm.latitude, trm.longitude, trm.altitude from respi_fmnr_species sp join respi_tree_measurement trm on trm.fmnr_species_id=sp.id;'''
query2 = '''select mt.id as microcatchment_id, mt.scientific_name, mt.local_name, im.latitude, im.longitude, im.altitude from respi_microcatchment_trees mt join respi_individual_mirocatchment im on im.id=mt.individual_mirocatchment_id;'''

# Execute the query and fetch the results
con = engine.connect()
df = pd.read_sql_query(query, con)
df1 = pd.read_sql_query(query1, con)
df2 = pd.read_sql_query(query2, con)

# Close the connection
con.close()
engine.dispose()

end_time = time.time()
elapsed_time = start_time - end_time
print(f'It took {elapsed_time} seconds to Query and get dataframes')
```

```python
import re

def clean_species(species_df: pd.DataFrame, scientific_name_col: str, local_name_col: str):
    """
    Clean species data and create a unique species dataframe.

    Parameters
    ----------
    species_df : pandas.DataFrame
        Dataframe containing species data.
    scientific_name_col : str
        Name of the column containing the scientific name.
    local_name_col : str
        Name of the column containing the local name.

    Returns
    -------
    cleaned_species_df : pandas.DataFrame
        Dataframe with cleaned species data.
    unique_scientific : pandas.DataFrame
        Dataframe with unique species.
    """

    species_df['scientificName'] = np.where(
        species_df[scientific_name_col].isna() | species_df[scientific_name_col].str.lower().eq('unknown'), # Check for 'unknown' and replace with local name
        species_df[local_name_col], # Use local name if scientific name is missing
        species_df[scientific_name_col] # Use scientific name if available
    )
    print("step 1.1: scientificName column created based on presense/absence of scientific name")
    # Create nameSource column based on conditions
    species_df['nameSource'] = np.select(
        [
            species_df[scientific_name_col].isna(),
            species_df[scientific_name_col].str.lower().eq('unknown'),
            species_df[scientific_name_col].notna()
        ],
        [
            'local_name (replaced NA)',
            'local_name (replaced Unknown)',
            'scientific_name'
        ],
        default='unknown_source'
    )
    print("step 1.2: nameSource column created based on source of name")
    
    # clean non-alphabetical characters
    species_df['scientificName'] = species_df['scientificName'].apply(
        lambda x: re.sub(r'[^a-zA-Z ]', '', str(x))
    )
    print("step 1.3: scientificName column cleaned of non-alphabetical characters")
    
    # trim whitespace
    species_df['scientificName'] = species_df['scientificName'].str.strip()
    print("step 1.4: scientificName column whitespace trimmed")


    # replace string "nan", "NaN", or empty strings with actual NaN
    species_df['scientificName'] = species_df['scientificName'].replace(
        ['nan', 'NaN', '', ' '], np.nan, regex=False
    )

    # Create unique species dataframe
    #unique_scientific = species_df[['scientificName','country']].drop_duplicates(subset='scientificName').dropna(subset='scientificName')
    
    # unique_scientific = species_df['scientificName'].unique()
    # unique_scientific = pd.DataFrame({'scientificName': unique_scientific})

    print("step 1.5: Unique species dataframe created")
    print("------------------------cleaning done------------------------")
    return species_df
```
