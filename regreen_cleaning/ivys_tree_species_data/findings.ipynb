# v1 data findings


```python
import pandas as pd
import os
import shapely.wkt


wd = os.path.join(os.path.expanduser('~'), 'Documents', 'projects', 'species_matching', 'python_scripts')

os.chdir(wd)

from sm_methods import clean_species

dir = os.path.join(os.path.expanduser('~'), 'Documents', 'projects', 'species_matching', 'data', 'Regreening_data')

cleaned_geom = pd.read_csv(os.path.join(dir, 'cleaned_geom_data.csv'))
#raw_source = pd.read_csv(os.path.join(dir, 'Regreening_all_raw_geom.csv))

```

## clean the raw_geom data


```python
dff = pd.read_csv(os.path.join(dir, 'Regreening_all_raw_geom.csv'))

clean_df, _ = clean_species(dff, 'species_name', 'local_name')

#dff.columns

clean_df = clean_df[['plotID', 'species_name', 'local_name', 'scientificName', 'nameSource', 'geometry']]

clean_df
```

# get the centroid 

```python
clean_df = clean_df.assign(
    location=lambda clean_df: clean_df['geometry'].apply(lambda x: shapely.wkt.loads(x).centroid)
)
clean_df 
```

 ## 54130 species names in raw data

```python
clean_df['scientificName'].count()
```

```python
clean_df = clean_df[['plotID', 'location']]
```

# clean the `clean_source data`

```python
dff = pd.read_csv(os.path.join(dir, 'Regreening_all_cleaned_2.csv'))
clean_dff, _ = clean_species(dff, 'scientific_name', 'local_name')

cleaned_source = clean_dff[['plotID', 'scientific_name', 'local_name', 'scientificName', 'nameSource']]

```
