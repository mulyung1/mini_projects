## TP

```python
tp = pd.read_csv('data_tp.csv')
```

```python
tp.columns
```

```python
filterd_tp = tp[['Tree Scientific Name', 'Tree Local Name', 'Total Trees Planted', 'Total Trees Survived', 'Tree Usage', 'Other Tree Usages']]
filterd_tp
```

```python
filterd_tp['has_fruits'] = filterd_tp['Tree Usage'].str.contains(
    'FRUITS', case=False, na=False
)
filterd_tp['country'] = 'Kenya'
filterd_tp
```

```python
filterd_tp.to_csv('tp_fruits.csv', index=False)
```

```python
cleaned_tp_fuits = pd.read_csv('tp_cleaned_fruits.csv')

unique_species = cleaned_tp_fuits.drop_duplicates(subset=['Tree Scientific Name'], keep='first')
unique_species = unique_species[['Tree Scientific Name', 'Tree Local Name', 'Total Trees Planted', 'Total Trees Survived', 'Tree Usage', 'Other Tree Usages', 'has_fruits']]
unique_species
```

```python
unique_species.to_csv('tp_unique_species.csv', index=False)
```
